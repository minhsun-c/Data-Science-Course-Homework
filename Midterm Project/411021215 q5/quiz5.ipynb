```python
import requests
url = "https://news.ycombinator.com"
html = requests.get(url)
html.encoding = "utf-8"

from bs4 import BeautifulSoup
sp = BeautifulSoup(html.text, 'lxml')
```

```python
row1 = sp.find_all(class_ = "titleline")
titles = []
hrefs = []

for row in row1:
    data = row.find_all("a")
    if len(data) == 2:
        titles.append(data[0].text)
        hrefs.append(data[1].text)
    else:
        titles.append(data[0].text)
        hrefs.append("None")
```

```python
scores = sp.find_all(class_ = "subtext")
for i in range(len(scores)):
    temp = scores[i].find(class_ = "score")
    if temp == None:
        scores[i] = "None"
    else:
        scores[i] = temp.text.replace(" points", "")
```

```python
import pandas as pd

dic = {
    "Title" : titles,
    "Score" : scores,
    "Href"  : hrefs
}

df = pd.DataFrame(dic)
df
```
