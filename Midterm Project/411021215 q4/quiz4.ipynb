```python
from bs4 import BeautifulSoup
import re

url = "/Users/min-hsunchang/Downloads/World_Population.html"
page = open(url)
soup = BeautifulSoup(page.read())
```

```python
trs = soup.find_all("tr")[1:]
```

```python
data = [i for i in trs][1:]
table = []
for i in range(len(data)):
    tds = trs[i].find_all("td")
    temp = []
    for j in range(1, len(tds)):
        if j == 6:
            temp.append(float(tds[j].text.replace(",", "")))
        else:
            temp.append(tds[j].text)
    table.append(temp)
```

```python
table.sort(key = lambda x : x[5], reverse = True)
```

```python
import pandas as pd
title = [
    "Country", "2023 Population", "2022 Population", 
    "Area", "Land Area", "Density", 
    "Growth Rate", "World", "Rank"
]
df = pd.DataFrame(table, columns = title)
df
```
