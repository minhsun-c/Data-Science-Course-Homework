```python
import pandas as pd
db = pd.read_csv('data.csv')
db
```

```python
import math
scores = [[] for i in range(10)]
for d in db:
    scores[0].append(math.floor(float(d)))
for i in range(1, 10):
    row = db.iloc[i-1]
    for j in range(len(row)):
        scores[i].append(db.iloc[i-1][j])
```

```python
def getScore(l):
    return l[0] * 0.1 + l[1] * 0.1 + l[2] * 0.1 + l[3] * 0.1 + l[4] * 0.2 + l[5] * 0.2 + l[6] * 0.2
```

```python
sol = [[0 for i in range(4)] for j in range(10)]
for i in range(10):
    sol[i][0] = getScore(scores[i][:7])
    sol[i][1] = getScore(scores[i][7:14])
    sol[i][2] = getScore(scores[i][14:21])
    sol[i][3] = (sol[i][0] + sol[i][0] + sol[i][0] + sum(scores[21:31]) * 10) * 0.3
    if sol[i][3] > 100: sol[i][3] = 100
    sol[i][0] = round(sol[i][0])
    sol[i][1] = round(sol[i][1])
    sol[i][2] = round(sol[i][2])
    sol[i][3] = round(sol[i][3])
sol
```

```python
df = pd.DataFrame(sol)
df.to_csv("out.csv", encoding='utf-8', index=False, header=False)
```
