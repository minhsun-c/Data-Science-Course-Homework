```python
import requests
url = 'https://sys.ndhu.edu.tw/SA/XSL_ApplyRWD/ActApply.aspx?lang=en'
html = requests.get(url)
html.encoding="utf-8"

from bs4 import BeautifulSoup
sp = BeautifulSoup(html.text, 'lxml')
```

```python
titlel = sp.select("th")
titlel = [i.text.replace("\n", "") for i in titlel if i.text!='\n'][1:]
titlel
```

```python
# Activity Name
# BodyContent_gvActs_lblGv_act_name_0
actName = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_act_name_{i}")
    if a != None:
        actName.append(a.text)
```

```python
# Enroll Deadline
# BodyContent_gvActs_lblGv_reg_dt_0
enrollDeadline = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_reg_dt_{i}")
    if a != None:
        enrollDeadline.append(a.text)
```

```python
# Activity Date
# BodyContent_gvActs_lblGv_act_dt_0
actDate = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_act_dt_{i}")
    if a != None:
        actDate.append(a.text)
```

```python
# Certified? /Hours
# BodyContent_gvActs_lblGv_xsl_check_0
certi = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_xsl_check_{i}")
    if a != None:
        a = a.text.split("/")
        if (len(a) == 2):
            s = (a[0] + a[1])
        else:
            s = (a[0])
        certi.append(s)
```

```python
# Quota
# BodyContent_gvActs_lblGv_reg_num_0
quota = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_reg_num_{i}")
    if a != None:
        a = a.text.split()
        quota.append("{:2} / {:2} People".format(a[0], a[2]))
```

```python
# Organizer
# BodyContent_gvActs_lblGv_act_unit_0
org = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_act_unit_{i}")
    if a != None:
        org.append(a.text)
```

```python
# Participant
# BodyContent_gvActs_lblGv_participant_0
part = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_participant_{i}")
    if a != None:
        part.append(a.text)
```

```python
# Status
# BodyContent_gvActs_lblGv_status_0
status = []
for i in range(20):
    a = sp.find(id=f"BodyContent_gvActs_lblGv_status_{i}")
    if a != None:
        status.append(a.text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    titlel[0]: status,
    titlel[1]: certi,
    titlel[2]: part,
    titlel[3]: actName,
    titlel[4]: enrollDeadline,
    titlel[5]: actDate,
    titlel[6]: quota,
    titlel[7]: org
})
```
