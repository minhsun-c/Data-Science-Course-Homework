```python
import requests
url = 'https://huggingface.co/models'
html = requests.get(url)
html.encoding="utf-8"

from bs4 import BeautifulSoup
sp = BeautifulSoup(html.text, 'lxml')
```

```python
articles = sp.find_all("article")
for i in range(len(articles)):
    art = articles[i].text.replace("\n", "#").replace("\t", "#").replace("â€¢", "#").replace("Updated", "#").split("#")
    articles[i] = [a for a in art if len(a) > 0 and a != " "]
```

```python
title = [
    "Project",
    "Type",
    "Updated",
    "Downloads",
    "Likes"
]
```

```python
sol = []
```

```python
for art in articles:
    dic = {}
    if len(art) == 4:
        dic[title[0]] = art[0]
        dic[title[1]] = art[1]
        dic[title[2]] = art[2]
        dic[title[3]] = "None"
        dic[title[4]] = art[3]
    elif len(art) == 5:
        dic[title[0]] = art[0]
        dic[title[1]] = art[1]
        dic[title[2]] = art[2]
        dic[title[3]] = art[3]
        dic[title[4]] = art[4]
    elif len(art) == 3:
        dic[title[0]] = art[0]
        dic[title[1]] = "None"
        dic[title[2]] = art[1]
        dic[title[3]] = "None"
        dic[title[4]] = art[2]
    sol.append(dic)
```
